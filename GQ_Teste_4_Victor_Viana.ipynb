```python
# importando pacotes
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns 
import numpy as np
# supress warnings
import warnings
warnings.filterwarnings('ignore')
```

```python
# criando serie com os dados fornecidos
df = pd.Series( 
        [17,22,19,26,13,10,21,14,11,14,
         25,15,15,16,21,20,13,17,15,14,
         13,8,22,18,15,25,13,14,11,15,
         12,15,10,21,13,16,15,7,14,14,
         10,19,18,15,20,13,27,18,14,15,
         26,15,16,14,17,11,13,20,12,15,
         18,19,18,21,14,16,13,10,15,24,
         14,21,13,8,25,18,14,16,20,9,
         13,18,12,14,17,13,17,19,20,13,
         13,11,20,7,23,19,21,18,26,12,
         19,14,18,27])
```



> **Como informa no histograma abaixo 90% das aprovações de crédito foram feitas em menos de 22 horas. Desse modo, o tempo de concessão de crédito foi efetivo.**



```python
# print de 90 percentil
print("90% das aprovações de crédito duraram menos de {:.2f} horas".format(df.quantile(.90)))

#plot do grafico
fig, ax = plt.subplots(figsize=(9,6))
sns.set_style("darkgrid")

# plotar dados com linhas de media e percentil 90
sns.distplot(df,kde=True,ax=ax,label='Horas')
plt.axvline(x=df.mean(),linestyle='--',color='g',ymax= 0.95,label='Média')
plt.axvline(x=df.quantile(.90),linestyle='--',color='r',ymax= 0.95,label='Percentil 90')

# personalizacao de layout e titulo
ax.set_title(r'Histograma - $\mu=16.22$, $\sigma=4.55$,',fontsize=13)
ax.legend()
plt.tight_layout()
```
